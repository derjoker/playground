```python
%matplotlib inline
import matplotlib.pyplot as plt
import json

data = json.load(open("data.json"))
data = data["data"]
print("total", len(data))
data = data[0:60]
# data = data[0:180]
# data = data[0:250]
# data = data[0:500]
data.reverse()
date = [item[0] for item in data]
print("from", date[0], "to", date[-1])
```

```python
# 融资买入（亿元）
rzBuy = [round(item[4]/100000000, 2) for item in data]
# 融券卖出量（亿股）
rqSaleCount = [round(item[7]/100000000, 2) for item in data]

fig = plt.figure()
axRzBuy = fig.subplots()
axRqSaleCount = axRzBuy.twinx()

axRzBuy.plot(date, rzBuy, 'r')
axRqSaleCount.plot(date, rqSaleCount, 'g')

plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import json

data = json.load(open("data.json"))
data = data["data"]
print("total", len(data))
# data = data[0:180]
# data = data[0:250]
data = data[0:500]
data.reverse()
date = [item[0] for item in data]
print("from", date[0], "to", date[-1])
```

```python
# 
rzBuy = [round(item[4]/100000000, 2) for item in data]
plt.plot(date, rzBuy)
plt.show()
```

```python
# 
rqSaleCount = [round(item[7]/100000000, 2) for item in data]
plt.plot(date, rqSaleCount)
plt.show()
```
